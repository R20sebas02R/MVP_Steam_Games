```python
import pandas as pd
import gzip
import ast
import json
```

```python
def abrir_comprimido(archivo):
    data_list = []
    with gzip.open(archivo, 'rb') as file:
        for line in file:
            decode_line = line.decode('utf-8')
            data_list.append(ast.literal_eval(decode_line.strip()))
        return data_list
```

```python
# Función para extraer los valores de 'item_id' de una lista de diccionarios
def extract_item_ids(review_list):
    return [diccionario['item_id'] for diccionario in review_list if 'item_id' in diccionario]
```

```python
# Crea un diccionario para almacenar las sumas acumulativas por género
sumas_por_genero = {genero: 0 for genero in lista_de_generos}
user_items_subset = user_items[['items']]
# Itera a través de las filas de 'users_items'
for i, row in user_items_subset.iterrows():
    items = row['items']
    print(i)

    for item in items:
        playtime_forever = item.get('playtime_forever')
        item_id = item.get('item_id')

        if playtime_forever is not None and item_id is not None:
            playtime_forever = int(playtime_forever)
            item_id = int(item_id)
            # Filtra el DataFrame 'steam_games' por 'item_id'
            generos_del_juego = steam_games.loc[steam_games['id'] == item_id, 'genres']

            # Verifica si se encontraron juegos con el 'item_id' deseado
            if not generos_del_juego.empty:
                genres_item = generos_del_juego.tolist()[0]
                # Expandir la lista de géneros y luego iterar sobre ellos
                for genero in genres_item:
                    if genero in sumas_por_genero:
                        sumas_por_genero[genero] = sumas_por_genero[genero] + playtime_forever

# El diccionario 'sumas_por_genero' ahora contiene las sumas acumulativas de 'playtime_forever'
# solo para los géneros a los que los juegos pertenecen.

diccionario_ordenado = dict(sorted(sumas_por_genero.items(), key=lambda x: x[1], reverse=True))
```

```python
diccionario_ordenado
```

```python
nombre_archivo = '/content/drive/My Drive/DataOutput_mvp_steam/generos_ranking.json'

# Guarda el diccionario en un archivo JSON
with open(nombre_archivo, 'w') as archivo:
    json.dump(diccionario_ordenado, archivo)
```

      Crear 'generos_usuarios.json'

```python
# Crear un diccionario de géneros únicos con diccionarios anidados
diccionario_generos_unicos = {genero: {} for genero in lista_de_generos}
user_items_subset = user_items[['user_id', 'user_url', 'items']]
# Iterar a través de las filas de 'users_items'
for i, row in user_items_subset.iterrows():
    print(i)
    user_id = row['user_id']
    user_url = row['user_url']
    items = row['items']

    for item in items:
        playtime_forever = item.get('playtime_forever')
        item_id = item.get('item_id')


        if playtime_forever is not None and item_id is not None:
            playtime_forever = int(playtime_forever)
            item_id = int(item_id)
            # Filtrar el DataFrame 'steam_games' por 'item_id'
            generos_del_juego = steam_games.loc[steam_games['id'] == item_id, 'genres']
            # Verifica si se encontraron juegos con el 'item_id' deseado
            if not generos_del_juego.empty:
                genres_item = generos_del_juego.tolist()[0]
                # Iterar a través de los géneros del juego y actualizar el diccionario
                for genero in genres_item:
                    if user_id in diccionario_generos_unicos[genero]:
                        diccionario_generos_unicos[genero][user_id][1] += playtime_forever
                    else:
                        diccionario_generos_unicos[genero][user_id] = [user_url, playtime_forever]
```
